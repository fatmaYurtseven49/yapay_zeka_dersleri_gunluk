```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x=np.arange(20)
y=np.random.normal(10,1,20)
z=np.random.normal(10,2,20)
k=np.random.normal(10,3,20)
```

```python
plt.plot(x,y,color='r')
plt.plot(x,z,color='y')
plt.plot(x,k);
```

```python
fig=plt.figure(figsize=(10,15))
ax=fig.add_subplot(311)
ax.set_title("x in y e gore degisimi")
ax.set_xlabel("x verileri ")
ax.plot(x,y)

bx=fig.add_subplot(312)
bx.set_title("z in x e gore degisimi")
bx.plot(x,z,color='red')
cx=fig.add_subplot(313)
cx.set_title("k in x e gore degisimi")
cx.plot(x,k)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['day']=(df['Date']).dt.day
df['month']=(df['Date']).dt.month
df['year']=(df['Date']).dt.year
```

```python
df['month_name']=df['Date'].dt.month_name()
df['weekday']=df['Date'].dt.day_name()
```

```python
df["Time"]=pd.to_datetime(df['Time'])
```

```python
df['Hour']=(df['Time']).dt.hour
```
