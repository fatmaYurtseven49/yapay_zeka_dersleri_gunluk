```python
import pandas as pd
```

```python
yelp=pd.read_csv("yelp.csv")
```

```python
yelp.head()
```

```python
yelp.info()
```

```python
yelp.head()
```

```python
yelp['text']=yelp['text'].str.lower()
#harfleri kucuk harfe cevir
```

```python
yelp['text']=yelp['text'].str.replace('[^\w\s]','')
#noktalama isaretlerini kaldirdiktik
```

```python
yelp['text']=yelp['text'].str.replace('\d+','')#rakamlari kaldirdik
```

```python
yelp['text']=yelp['text'].str.replace('\n','').replace('\r','')
#enter ve yeni satir kodlarini kaldirdik
```

```python
yalp_ei_ek=yelp[(yelp.stars==5) | (yelp.stars==1) ] 
```

```python
yalp_ei_ek.reset_index(drop=True,inplace=True)
```

```python
yalp_ei_ek
```

```python
x=yalp_ei_ek.text
y=yalp_ei_ek.stars
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
vect=CountVectorizer(lowercase=True,stop_words='english')
x_yeni=vect.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_yeni,y,random_state=42)
```

```python
tf=pd.DataFrame(vect.fit_transform(x_train).toarray(),columns=vect.get_feature_names())
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mb=MultinomialNB()
```

```python
mb.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
tahmin=mb.predict(x_test)
```

```python
accuracy_score(y_test,tahmin)
```
